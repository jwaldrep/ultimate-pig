```python
import statistics as st
import random
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pig_player import PigPlayer, Die
from pig_game import PigSolitaire
```

##Simulation

```python
def solitaire_trials(num_trials=1000, player=PigPlayer):
    scores = []
    for _ in range(num_trials):
        die = Die()
        the_player = player(die)
        game = PigSolitaire(the_player)
        scores.append(game.play())
    print('num_trials: {}  Mean: {}  Stdev: {}'.format(
            num_trials,st.mean(scores), st.stdev(scores)))
    return scores
```

```python
sol_10 = solitaire_trials(10)
sol_100 = solitaire_trials(100)
sol_1000 = solitaire_trials(1000)
sol_10000 = solitaire_trials(10000)
sol_100000 = solitaire_trials(100000)
sol_1000000 = solitaire_trials(1000000)
```

```python
def plot_y_hist(data, title = "Plot"):
    plt.hist(data, bins=20)
    plt.xlabel("score")
    plt.ylabel("frequency")
    plt.title(title)

    ymin, ymax = plt.ylim()
    mean = st.mean(data)
    stdev = st.stdev(data)
    plt.vlines(mean, ymin, ymax, color="red")
    plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, linestyles="dashed", color="red")
    plt.show()
```
